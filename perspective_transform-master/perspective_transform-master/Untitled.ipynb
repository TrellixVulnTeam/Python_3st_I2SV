```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./img/aa.jpg')

plt.imshow(img)
```

```python
src = np.array([
    [-30, 30],
    [460, -30],
    [520, 460],
    [30, 520]
], dtype=np.float32)

width = max(np.linalg.norm(src[0] - src[1]), np.linalg.norm(src[2] - src[3]))
height = max(np.linalg.norm(src[0] - src[3]), np.linalg.norm(src[1] - src[2]))

dst = np.array([
    [0, 0],
    [width, 0],
    [width, height],
    [0, height]
], dtype=np.float32)

M = cv2.getPerspectiveTransform(src, dst)
M
```

```python
warped = cv2.warpPerspective(img, M, (width, height))

plt.imshow(warped)
```
